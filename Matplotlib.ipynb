```python
'''
Scatter plot : A scatter plot is a type of data visualization that displays individual data points on a two-dimensional 
graph. Each point represents the values of two variables, and the position of the point is determined by the values of 
those variables. Scatter plots are useful for visually examining the relationship between two continuous variables.
'''
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
'''
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.
It allows you to plot multiple plots in a single figure, making it convenient for visualizing and 
comparing different datasets or aspects of your data.
'''

import matplotlib.pyplot as plt
import numpy as np

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, plot 1
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, plot 2
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, plot 3
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, plot 4
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

```python
'''
Bar plot : A bar plot is a type of data visualization that presents categorical data with rectangular bars. 
The lengths of the bars are proportional to the values they represent. Bar plots are useful for 
comparing the values of different categories or groups, making it easy to identify trends or variations in the data.
'''
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Profit by Company (Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.title('Profit by Company (Horizontal Bar Plot)')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```
